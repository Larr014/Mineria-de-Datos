```python
# prompt: Lee como dataframe el archivo csv

import pandas as pd

df = pd.read_csv('/content/titanic_procesado.csv')
print(df.head())
```

```python
# prompt: Crea un dataframe que solo guarde las columnas Pclass, Sex, Age , Survived y SibSp

new_df = df[['Pclass', 'Sex', 'Age', 'Survived', 'SibSp']]
print(new_df.head())
```

Separo las columnas que necesito para X e y

```python
X = new_df.drop('Survived', axis=1)
y = new_df['Survived']
```

```python
X
```

```python
y
```

Revisamos la columna Sex

```python
X['Sex'].unique()
```

Categorizar

```python
X['Sex'] = X['Sex'].map({'male': 0, 'female': 1})
X['Sex']
```

```python
X
```

```python
# prompt: aplica minmax scaler

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming X is your DataFrame
# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the numerical columns in X
numerical_cols = ['Age', 'SibSp', 'Pclass','Sex']  # Replace with the actual column names
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

print(X.head())
```

```python
# prompt: une X e y y guardalos en un archivo csv



# Combine X and y into a single DataFrame
combined_df = pd.concat([X, y], axis=1)

# Save the combined DataFrame to a CSV file
combined_df.to_csv('titanic_limpio_procesado.csv', index=False)
```
