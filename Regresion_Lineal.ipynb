# Probando Demografia

```python
import pandas as pd
df = pd.read_csv('/content/datosDemograficos.csv')
df.head()
```

```python
df1 = df.copy()
x = df1.drop(['Nombre.Pais','Codigo.Pais','CantPersonas.Internet','Grupo.Ingresos'], axis=1)
x
```

```python
df2 = df.copy()
y = df2.drop(['Nombre.Pais','Codigo.Pais','Tasa.Natalidad','Grupo.Ingresos'], axis=1)
y
```

```python
df.describe()
```

```python
# prompt: grafica x e y en un scatter y que se muestre la regresion lineal

import matplotlib.pyplot as plt
import numpy as np

# Calcular la regresión lineal
slope, intercept = np.polyfit(x['Tasa.Natalidad'], y['CantPersonas.Internet'], 1)

# Crear la gráfica de dispersión
plt.scatter(x['Tasa.Natalidad'], y['CantPersonas.Internet'])

# Agregar la línea de regresión
plt.plot(x['Tasa.Natalidad'], slope * x['Tasa.Natalidad'] + intercept, color='red')

# Etiquetas y título
plt.xlabel('Tasa de Natalidad')
plt.ylabel('Cantidad de Personas con Internet')
plt.title('Relación entre Tasa de Natalidad y Cantidad de Personas con Internet')

# Mostrar la gráfica
plt.show()
```

```python
# prompt: implementa un algoritmo en python que aplica regresion lineal usando scikitlearn

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Datos de ejemplo (reemplaza con tus propios datos)
#X = np.array([[1], [2], [3], [4], [5]])
#y = np.array([2, 4, 5, 4, 6])

# Divide los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Crea un modelo de regresión lineal
model = LinearRegression()

# Entrena el modelo con los datos de entrenamiento
model.fit(X_train, y_train)

# Realiza predicciones con los datos de prueba
y_pred = model.predict(X_test)

# Evalúa el modelo calculando el error cuadrático medio
mse = mean_squared_error(y_test, y_pred)

print("Error cuadrático medio:", mse)
mae = mean_absolute_error(y_test, y_pred)
print("Error absoluto medio:", mae)
r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R^2):", r2)

# Puedes usar el modelo para predecir nuevos valores
new_value = np.array([[6]])
prediction = model.predict(new_value)
print("Predicción para el nuevo valor:", prediction)
```
