```python
# prompt: Lee como dataframe el archivo csv

import pandas as pd

df = pd.read_csv('/content/dragon_ball_z.csv')
print(df.head())
```

```python
df
```

```python
# prompt: Crea un dataframe que solo guarde las columnas Pclass, Sex, Age , Survived y SibSp

new_df = df[['Race', 'Gender', 'Power Level', 'Speed']]
print(new_df.head())
```

```python
new_df=new_df.dropna()
new_df
```

Separo las columnas que necesito para X e y

```python
new_df['Race'] = new_df['Race'].astype('category')
new_df['Race'] = new_df['Race'].cat.codes
new_df['Gender'] = new_df['Gender'].astype('category')
new_df['Gender'] = new_df['Gender'].cat.codes
new_df['Speed'] = new_df['Speed'].astype('category')
new_df['Speed'] = new_df['Speed'].cat.codes
new_df
```

```python
X = new_df.drop('Race', axis=1)
y = new_df['Race']
```

Categorizar

```python
# prompt: aplica minmax scaler

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming X is your DataFrame
# Create a MinMaxScaler object
scaler = MinMaxScaler()

# Fit and transform the numerical columns in X
numerical_cols = ['Gender',"Power Level","Speed"]  # Replace with the actual column names
X[numerical_cols] = scaler.fit_transform(X[numerical_cols])

print(X.head())
```

```python
# prompt: une X e y y guardalos en un archivo csv



# Combine X and y into a single DataFrame
combined_df = pd.concat([X, y], axis=1)

# Save the combined DataFrame to a CSV file
combined_df.to_csv('dbz_limpio_procesado.csv', index=False)
```
